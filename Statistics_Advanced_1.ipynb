1.What is a random variable probability theory?
- A random variable in probability theory is a function that maps the outcomes of a random phenomenon to real numbers.

2.What are the types of random variables?
- The two main types of random variables are discrete and continuous

3.Explain the difference between discrete and continuous distributions.
- a discrete distribution represents variables with a finite or countable number of outcomes, while a continuous distribution models variables that can take any value within a given range

4.What is a binomial distribution, and how is it used in probability?
- A binomial distribution is a discrete probability distribution that models the number of "successes" in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure.

5.What is the standard normal distribution, and why is it important?
- The standard normal distribution is a specific version of the normal distribution with a mean of 0 and a standard deviation of 1.it provides a standardized way to compare different normal distributions by converting raw scores into z-scores, which indicate how many standard deviations a data point is from the mean.

6.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the distribution of sample means will be approximately normal, regardless of the original population's distribution, as long as the sample size is large enough. It is critical in statistics because it allows us to make inferences about a population using sample data, forming the basis for many statistical methods like confidence intervals and hypothesis testing, even when the original population is unknown or non-normal.

7.What is the significance of confidence intervals in statistical analysis?
- The significance of confidence intervals in statistical analysis is that they provide a range of values within which a population parameter is likely to lie, quantifying uncertainty and providing a more complete picture than a single point estimate.

8.What is the concept of expected value in a probability distribution?
- The expected value of a probability distribution is the weighted average of all possible outcomes, calculated by summing the product of each outcome and its probability.



```python
#9.Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt
mean = 50
std_dev = 5
num_samples = 1000

random_numbers = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

print(f"Generated {num_samples} random numbers from a normal distribution.")
print(f"Theoretical Mean: {mean}, Theoretical Standard Deviation: {std_dev}")
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

plt.hist(random_numbers, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

10.You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

- Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
  - apply the Central Limit Theorem by taking multiple random samples from the daily sales data, calculating the mean of each sample, and then using the distribution of these sample means to find a margin of error around the overall sample mean
   

```python
#10.Write the Python code to compute the mean sales and its confidence interval.

import numpy as np
from scipy.stats import norm

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
sales_array = np.array(daily_sales)
sample_mean = np.mean(sales_array)
print(f"Sample Mean: {sample_mean:.2f}")

sample_std = np.std(sales_array, ddof=1)
n = len(sales_array)
standard_error = sample_std / np.sqrt(n)

z_score = norm.ppf(0.975)
margin_of_error = z_score * standard_error

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Standard Error: {standard_error:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

```
