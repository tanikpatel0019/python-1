6. What are the differences between NumPy arrays and Python lists ?
- NumPy Array
   - Homogeneous
   - Stored in contiguous memory blocks, making them more efficient.
   - Much faster for numerical computations.
   - Comes with many built-in mathematical, linear algebra, and statistical functions.
- Python List
     - Heterogeneous
     - Stored as references (pointers) to objects, so memory usage is higher.
     - Slower for large-scale numerical operations.
     - Limited functionality; requires manual implementation.

17. Explain the difference between apply() and map() in Pandas ?
- apply()
   - Works on both Series and DataFrames.
   - For a Series: works like map() (applies function element-wise).
   - For a DataFrame: applies a function along an axis (axis=0 for columns, axis=1 for rows).
- map()
    - Works only on Pandas Series (one-dimensional).
    - Applies a function element-wise to each value in the Series.

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Salary": [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
mean_age = df["Age"].mean()
print("Mean Age:", mean_age)

mean_salary = df["Salary"].mean()
print("Mean Salary:", mean_salary)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]
y3 = [12, 18, 28, 33, 45]
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', marker='s')
plt.plot(x, y3, label='Line 3', color='green', marker='^')
plt.title("Multiple Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Salary": [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
filtered_df = df[df["Salary"] > 60000]
print("\nFiltered DataFrame (Salary > 60000):\n", filtered_df)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.DataFrame({
    "Age": np.random.normal(loc=30, scale=5, size=100)
})
plt.figure(figsize=(8,6))
sns.histplot(data["Age"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
#11. Perform matrix multiplication using NumPy.

import numpy as np
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])
result = np.dot(A, B)
print("Result using np.dot():\n", result)
```

```python
#13. Create a 3D scatter plot using Plotly.

import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
data = pd.DataFrame({
    "X": np.random.rand(50),
    "Y": np.random.rand(50),
    "Z": np.random.rand(50),
    "Category": np.random.choice(["A", "B"], size=50)
})
fig = px.scatter_3d(
    data,
    x="X",
    y="Y",
    z="Z",
    color="Category",
    size="X",
    symbol="Category",
    title="3D Scatter Plot Example"
)
fig.show()
```
