1.What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
- Descriptive
   - Purpose
       -describe and summarize a dataset.
   - Scope -A specific, limited dataset.
   - Key Tools -Measures like mean, median, mode, standard deviation, charts, and graphs.
   - Example -Calculating the average test score of 100 students in a single class.

- Inferential
   - Purpose -To make predictions and generalizations about a population based on a sample.
   - Scope -A larger population that the sample represents.
   - Key Tools -Hypothesis tests, confidence intervals, and probability.
   - Example -Using the 100 students' scores to estimate the average test score of all 10,000 students in the state

2.What is sampling in statistics? Explain the differences between random
and stratified sampling.
- simple random sampling treats the entire population as a single group, while stratified sampling first divides the population into distinct, non-overlapping subgroups (strata) based on shared characteristics and then takes a random sample from each subgroup.

3.Define mean, median, and mode. Explain why these measures of central
tendency are important.
- Mean
  - The arithmetic average of all data points. It is calculated by adding all the values in the dataset and dividing by the number of values.
- Median
  - The middle value in a dataset that has been ordered from least to greatest. It is not affected by outliers.
- Mode
  - The value that appears most frequently in the dataset. A dataset can have one mode, multiple modes, or no mode at all.
- Importance
  - Data summarization
  - Understanding distribution
  - Comparison
  - Informing decisions

4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
- Skewness
  - A measure of asymmetry. It tells you if data points are spread out more on one side of the mean than the other.
- Kurtosis
  - A measure of the "tailedness" or "peakedness" of a distribution compared to a normal distribution.  
- Positive skew implications
  - A positive skew implies the data is not symmetric.
  - Most of the data is concentrated on the left side of the distribution.
  - There is a tail of high-value data points extending to the right.
  - The mean is greater than the median, which is greater than the mode.
  - This can be seen in data like income, where a few very high earners pull the average up significantly from the majority of the population.



```python
#5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

from collections import Counter

def calculate_mean(numbers):
    """Calculates the mean of a list of numbers."""
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    """Calculates the median of a list of numbers."""
    if not numbers:
        return 0
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 1:
        return sorted_numbers[n // 2]
    else:
        mid1 = sorted_numbers[n // 2 - 1]
        mid2 = sorted_numbers[n // 2]
        return (mid1 + mid2) / 2

def calculate_mode(numbers):
    """Calculates the mode(s) of a list of numbers."""
    if not numbers:
        return []
    counts = Counter(numbers)
    max_count = max(counts.values())
    modes = [key for key, value in counts.items() if value == max_count]
    return modes

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = calculate_mean(numbers)
median_value = calculate_median(numbers)
mode_values = calculate_mode(numbers)

print(f"The given list of numbers is: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_values}")
```

```python
#6.Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

array_x = np.array(list_x)
array_y = np.array(list_y)

covariance_matrix = np.cov(array_x, array_y)
covariance = covariance_matrix[0, 1]

correlation_coefficient = np.corrcoef(array_x, array_y)[0, 1]

print(f"Covariance between list_x and list_y: {covariance}")
print(f"Correlation coefficient between list_x and list_y: {correlation_coefficient}")
```

```python
#7.Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Boxplot of Data')
plt.ylabel('Values')
plt.grid(True)
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

```python
#8.You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# ● Explain how you would use covariance and correlation to explore this relationship.
# ● Write Python code to compute the correlation between the two lists:
# advertising_spend = [200, 250, 300, 400, 500]
# daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

spend_array = np.array(advertising_spend)
sales_array = np.array(daily_sales)
correlation_matrix = np.corrcoef(spend_array, sales_array)
correlation_coefficient = correlation_matrix[0, 1]

print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print(f"\nThe correlation coefficient between advertising spend and daily sales is: {correlation_coefficient:.4f}")
```

```python
#9.Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
# ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
# ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
bins = np.arange(min(survey_scores), max(survey_scores) + 2) - 0.5

plt.figure(figsize=(8, 5))
plt.hist(survey_scores, bins=bins, color='skyblue', edgecolor='black')
plt.title('Distribution of Customer Satisfaction Survey Scores (1-10 Scale)')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')

plt.xticks(range(min(survey_scores), max(survey_scores) + 1))
plt.grid(axis='y', alpha=0.7)
plt.show()
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
std_dev = np.std(survey_scores)

print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score}")
print(f"Standard Deviation: {std_dev:.2f}")
```
